## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
cols = df.columns
colors = ['#000099', '#ffff00'] #specifying colors; yellow=missing, blue=not missing
sns.heatmap(df[cols].isna(), cmap=sns.color_palette(colors))
```

```python
# Try out a missing data percentage list! 
missing_data_percents = df.isna().mean() * 100
missing_data_percents
```

```python
# Plot a missing data histogram
missing_by_row = df.isna().sum(axis='columns')
missing_by_row.hist(bins=50)
```

Now that you have tried all 3 techniques for detecting missing data, did you find any? Reading through the article, which solution do you think will help you clean the data set and handle the missing data? Try out your solution below!

```python
# Handle any missing data in your dataframe.
# Drop all columns that have above 90% null values
df_above_90_null = df[df.columns[(df.isna().mean() * 100)>=90]]
print(df_above_90_null.columns)
print(df.shape)
df = df.drop(axis=1, columns=df_above_90_null.columns)
print(df.shape)
```

```python
# Investigate 0% null columns
df_zero_null = df[df.columns[(df.isna().mean() * 100)==0]]
print(df_zero_null.info())
```

```python
print(df.shape)
uni_columns = []
for col in df_zero_null:
  print(f"{col}: {len(df_zero_null[col].unique())}")
  if len(df_zero_null[col].unique()) == 1:
        uni_columns.append(col)
df = df.drop(axis=1, columns=uni_columns)
print(df.shape)
```
